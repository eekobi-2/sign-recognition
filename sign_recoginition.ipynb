```python
def mediapipe_detections(image, model):
  img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  img.flags.writable = False
  img, results = model.process(img)
  img.flags.writable = False
  img = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

  return img, results
```

```python
def draw_landmarks(image, results):
    draw.draw_landmarks(image, results.face_landmarks, holi.FACE_CONNECTIONS) 
    draw.draw_landmarks(image, results.pose_landmarks, holi.POSE_CONNECTIONS) 
    draw.draw_landmarks(image, results.left_hand_landmarks, holi.HAND_CONNECTIONS) 
    draw.draw_landmarks(image, results.right_hand_landmarks, holi.HAND_CONNECTIONS) 
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened() :
  ret, frame = cap.read()

  img, res = mediapipe_detection(frame, holistics)
  
  draw_landmarks(img, res)


  cv2_imshow(img)
  if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
